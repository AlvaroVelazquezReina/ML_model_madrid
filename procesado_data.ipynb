### CALENDAR

```python
## a calendar le quitamos las labels minimum-maximum nights y adjusted price
```

```python
calendar = calendar.drop(['adjusted_price', 'minimum_nights', 'maximum_nights'], axis=1)
calendar.head()
```

```python
calendar.isnull().sum()
```

```python
## ahora le hacemos label encoding a avaliable: 0 o 1 en vez de True False
```

```python
from sklearn import preprocessing
```

```python
label_encoding = preprocessing.LabelEncoder()
calendar['available'] = label_encoding.fit_transform(calendar['available'].astype(str))

calendar.head()
```

```python
calendar.shape
```

```python
## guardamos calendar
calendar = pd.DataFrame(calendar)
calendar.to_csv('data_procesada/calendar.csv')
```

### Listings

```python
listings.shape
```

```python
listings.isnull().sum()
```

```python
listings.head()
```

```python
listings = listings.drop(['id','name', 'host_id', 'host_name','last_review', 
                          'reviews_per_month', 'calculated_host_listings_count'], axis=1)
listings.head()
```

```python
listings.isnull().sum()
```

```python
## hay que hacer label encoding a: 'neighbourhood_group', 'neighbourhood' y 'roomprice'
```

```python
label_encoding = preprocessing.LabelEncoder()
listings['neighbourhood_group'] = label_encoding.fit_transform(listings['neighbourhood_group'].astype(str))

listings.head()
```

```python
label_encoding = preprocessing.LabelEncoder()
listings['neighbourhood'] = label_encoding.fit_transform(listings['neighbourhood'].astype(str))

listings.head()
```

```python
label_encoding = preprocessing.LabelEncoder()
listings['room_type'] = label_encoding.fit_transform(listings['room_type'].astype(str))

listings.head()
```

```python
listings.isnull().sum()
```

```python
listings.shape
```

```python
## guardamos listings
listings = pd.DataFrame(listings)
listings.to_csv('data_procesada/listings.csv')
```
